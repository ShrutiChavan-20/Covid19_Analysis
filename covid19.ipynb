```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
print("\nMissing values:\n", df.isnull().sum())
```

```python
print("\nSummary Stats:")
print(df.describe())
```

```python
region_cases = df.groupby("RegionName")['TotalPositiveCases'].max().sort_values(ascending=False)
print("\nTotal Positive Cases by Region:\n", region_cases)
```

```python
# ========================
# Visualization 1: Top 10 Regions by Cases
# ========================
plt.figure(figsize=(10,5))
sns.barplot(x=region_cases.head(10).values, y=region_cases.head(10).index, palette="Reds_r")
plt.title("Top 10 Regions by Total COVID-19 Cases", fontsize=14, fontweight='bold')
plt.xlabel("Total Positive Cases")
plt.ylabel("Region")
plt.show()
```

```python
# Visualization 2: National Trend Over Time
# ========================
national_trend = df.groupby("Date")[['TotalPositiveCases','Deaths','Recovered']].sum()

plt.figure(figsize=(12,6))
plt.plot(national_trend.index, national_trend['TotalPositiveCases'], label='Total Positive Cases')
plt.plot(national_trend.index, national_trend['Deaths'], label='Deaths')
plt.plot(national_trend.index, national_trend['Recovered'], label='Recovered')
plt.legend()
plt.title("COVID-19 National Trend in Italy", fontsize=14, fontweight='bold')
plt.xlabel("Date")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```

```python
# ========================
# Visualization 3: Heatmap of Correlation
# ========================
plt.figure(figsize=(10,6))
sns.heatmap(df[['HospitalizedPatients', 'IntensiveCarePatients', 
                'TotalHospitalizedPatients', 'HomeConfinement',
                'CurrentPositiveCases', 'NewPositiveCases', 'Recovered', 'Deaths']].corr(), 
            annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Key Metrics", fontsize=14, fontweight='bold')
plt.show()
```

```python
# 2ï¸âƒ£ Daily New Cases in Italy
national_daily_cases = df.groupby("Date")['NewPositiveCases'].sum()

plt.figure(figsize=(12,6))
sns.lineplot(x=national_daily_cases.index, y=national_daily_cases.values, marker="o")
plt.title("Daily New Positive Cases in Italy", fontsize=14, fontweight='bold')
plt.xlabel("Date")
plt.ylabel("New Positive Cases")
plt.grid(True)
plt.show()
```

```python

# 3ï¸âƒ£ Cumulative Cases, Deaths, and Recoveries
national_totals = df.groupby("Date")[['TotalPositiveCases','Deaths','Recovered']].sum()

plt.figure(figsize=(12,6))
for col in national_totals.columns:
    plt.plot(national_totals.index, national_totals[col], label=col)
plt.legend()
plt.title("Cumulative COVID-19 Trends in Italy", fontsize=14, fontweight='bold')
plt.xlabel("Date")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```

```python
# 5ï¸âƒ£ Hospitalized vs. Intensive Care Scatter Plot
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="HospitalizedPatients", y="IntensiveCarePatients", hue="RegionName", alpha=0.7)
plt.title("Hospitalized vs Intensive Care Patients", fontsize=14, fontweight='bold')
plt.xlabel("Hospitalized Patients")
plt.ylabel("Intensive Care Patients")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# 6ï¸âƒ£ Death Rate by Region
region_death_rate = (df.groupby("RegionName")['Deaths'].max() / 
                     df.groupby("RegionName")['TotalPositiveCases'].max()) * 100
region_death_rate = region_death_rate.sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=region_death_rate.values, y=region_death_rate.index, palette="coolwarm")
plt.title("Death Rate (%) by Region", fontsize=14, fontweight='bold')
plt.xlabel("Death Rate (%)")
plt.ylabel("Region")
plt.show()
```

```python
# 8ï¸âƒ£ Active Cases vs. Deaths (Bubble Size = Recovered)
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="CurrentPositiveCases", y="Deaths", size="Recovered",
                hue="RegionName", alpha=0.6, sizes=(20, 400))
plt.title("Active Cases vs Deaths (Bubble size = Recovered)", fontsize=14, fontweight='bold')
plt.xlabel("Active Cases")
plt.ylabel("Deaths")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
